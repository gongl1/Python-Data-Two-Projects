** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
# Empty dataframe
returns = pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
# dataframe.pct_change() function calculates the percentage change between the current and a prior element. This function by default calculates the percentage change from the immediately previous row.

for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
BAC['Close'].loc['2008-01-01':'2009-01-01']
```

```python
BAC['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean()
```

```python
# Use .rolling(window=30).mean()

plt.figure(figsize=(12,6))
BAC['Close'].ix['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 Day Avg')
BAC['Close'].ix['2008-01-01':'2009-01-01'].plot(label='BAC CLOSE')
plt.legend()
```

** Create a heatmap of the correlation between the stocks Close Price.**

```python
# stocks Close Price

bank_stocks.xs(key='Close',axis=1,level='Stock Info')
```

```python
# Correct color legend!

sns.heatmap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True,cmap=sns.cubehelix_palette(light=.95, as_cmap=True))
```

** Optional: Use seaborn's clustermap to cluster the correlations together:**

```python
sns.clustermap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True,cmap=sns.cubehelix_palette(light=.95, as_cmap=True))
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr()
```

```python
# Use plotly

close_corr = bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr()
close_corr.iplot(kind='heatmap',colorscale='rdylbu')
```
